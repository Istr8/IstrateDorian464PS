```python
import numpy as np
from matplotlib import pyplot as plt
```

Exercitiul 1

```python
# Parametrii semnalului sinusoidal
amplitudine = 2
frecventa = 0.5  # Hz
faza = np.pi / 4  # Radiani

# Definirea timpului
timp = np.linspace(0, 10, 1000)

# Generarea semnalului sinusoidal
semnal_sinus = amplitudine * np.sin(2 * np.pi * frecventa * timp + faza)
semnal_cosinus = amplitudine * np.cos(2 * np.pi * frecventa * timp + faza)

# Afisarea graficului
plt.figure(figsize=(12, 6))
# Subplotul pentru semnalul sinusoidal
plt.subplot(2, 1, 1)
plt.plot(timp, semnal_sinus, label='Semnal sinusoidal', color='b')
plt.title('Semnal sinusoidal')
plt.xlabel('Timp')
plt.ylabel('Amplitudine')
plt.legend()

# Afisarea graficului
plt.tight_layout()
plt.show()
```

Exercitiul 2

```python
# Parametrii semnalului
amplitudine = 1
frecventa = 5  # Hz
faze = [0, np.pi/4, np.pi/2, 3*np.pi/4]  # diferite valori pentru faza

# Vectorul de timp
timp = np.linspace(0, 2 * np.pi, 1000)

# Generarea semnalului sinusoidal i adugarea de zgomot
plt.figure(figsize=(10, 6))
for index, faza in enumerate(faze):
    semnal = amplitudine * np.sin(2 * np.pi * frecventa * timp + faza)

    # Adugarea zgomotului
    snr_values = [0.1, 1, 10, 100]
    for snr in snr_values:
        gamma = np.sqrt(np.linalg.norm(semnal) ** 2 / (snr * np.linalg.norm(np.random.normal(size=len(semnal))) ** 2))
        zgomot = gamma * np.random.normal(size=len(semnal))
        semnal_zgomot = semnal + zgomot

        plt.plot(timp, semnal_zgomot, label=f'Faza: {faza}, SNR: {snr}')

# Setarea detaliilor pentru grafic
plt.title('Semnal sinusoidal cu zgomot la diferite SNR-uri i faze')
plt.xlabel('Timp')
plt.ylabel('Amplitudine')
plt.legend()
plt.show()
```

Exercitiul 6

```python
# Definim frecvena de eantionare i timpul
fs = 100  # Frecvena de eantionare
t = np.linspace(0, 1, fs, endpoint=False)

# Definim frecvenele fundamentale
frecvente = [fs / 2, fs / 4, 0]

# Generm cele trei semnale sinusoidale
semnale = [np.sin(2 * np.pi * f * t) for f in frecvente]

# Afim cele trei semnale
plt.figure(figsize=(10, 6))

for i in range(3):
    plt.subplot(3, 1, i + 1)
    plt.plot(t, semnale[i], label=f'f = {frecvente[i]} Hz')
    plt.title(f'Semnal sinusoidal cu frecvena {frecvente[i]} Hz')
    plt.xlabel('Timp')
    plt.ylabel('Amplitudine')
    plt.legend()

plt.tight_layout()
plt.show()
```

Exercitiul 7

```python
# Definim frecvena de eantionare i timpul
fs = 1000  # Frecvena de eantionare iniial
t = np.linspace(0, 1, fs, endpoint=False)

# Generm semnalul sinusoidal iniial
initial_signal = np.sin(2 * np.pi * 5 * t)  # Presupunem o frecven de 5 Hz

# Decimm semnalul
decimated_signal_1 = initial_signal[::4]  # Decimare ncepnd cu primul element
decimated_signal_2 = initial_signal[1::4]  # Decimare ncepnd cu al doilea element

# Afim graficele
plt.figure(figsize=(12, 6))

# Afim semnalul iniial
plt.subplot(3, 1, 1)
plt.plot(t, initial_signal, label='Initial Signal', color='b')
plt.title('Semnalul iniial i semnalele decimate')
plt.xlabel('Timp')
plt.ylabel('Amplitudine')
plt.legend()

# Afim primul semnal decimat
plt.subplot(3, 1, 2)
t_decimated = np.linspace(0, 1, len(decimated_signal_1), endpoint=False)
plt.plot(t_decimated, decimated_signal_1, label='Decimated Signal (Start from 1st element)', color='r')
plt.xlabel('Timp')
plt.ylabel('Amplitudine')
plt.legend()

# Afim al doilea semnal decimat
plt.subplot(3, 1, 3)
t_decimated = np.linspace(0, 1, len(decimated_signal_2), endpoint=False)
plt.plot(t_decimated, decimated_signal_2, label='Decimated Signal (Start from 2nd element)', color='g')
plt.xlabel('Timp')
plt.ylabel('Amplitudine')
plt.legend()

plt.tight_layout()
plt.show()
```

(a) n primul grafic, se observ c semnalul iniial are o frecven de eantionare de 1000 Hz, n timp ce semnalul decimat are o frecven de eantionare de 250 Hz. Astfel, se observ o pierdere semnificativ de informaii i o distorsiune a semnalului original.

(b) n al doilea grafic, semnalul decimat ncepnd cu al doilea element arat mai similar cu semnalul iniial dect n cazul decimrii ncepnd cu primul element. Acest lucru se datoreaz faptului c semnalul iniial a fost decimat iniial cu o frecven de eantionare de 250 Hz, iar acest proces a eliminat o parte din informaiile semnalului original.
