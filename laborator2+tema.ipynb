```python
import numpy as np
from matplotlib import pyplot as plt
```

Exercitiul 1

```python
# Parametrii semnalului sinusoidal
amplitudine = 2
frecventa = 0.5  # Hz
faza = np.pi / 4  # Radiani

# Definirea timpului
timp = np.linspace(0, 10, 1000)

# Generarea semnalului sinusoidal
semnal_sinus = amplitudine * np.sin(2 * np.pi * frecventa * timp + faza)
semnal_cosinus = amplitudine * np.cos(2 * np.pi * frecventa * timp + faza)

# Afisarea graficului
plt.figure(figsize=(12, 6))
# Subplotul pentru semnalul sinusoidal
plt.subplot(2, 1, 1)
plt.plot(timp, semnal_sinus, label='Semnal sinusoidal', color='b')
plt.title('Semnal sinusoidal')
plt.xlabel('Timp')
plt.ylabel('Amplitudine')
plt.legend()

# Afisarea graficului
plt.tight_layout()
plt.show()
```

Exercitiul 2

```python
# Parametrii semnalului
amplitudine = 1
frecventa = 5  # Hz
faze = [0, np.pi/4, np.pi/2, 3*np.pi/4]  # diferite valori pentru faza

# Vectorul de timp
timp = np.linspace(0, 2 * np.pi, 1000)

# Generarea semnalului sinusoidal și adăugarea de zgomot
plt.figure(figsize=(10, 6))
for index, faza in enumerate(faze):
    semnal = amplitudine * np.sin(2 * np.pi * frecventa * timp + faza)

    # Adăugarea zgomotului
    snr_values = [0.1, 1, 10, 100]
    for snr in snr_values:
        gamma = np.sqrt(np.linalg.norm(semnal) ** 2 / (snr * np.linalg.norm(np.random.normal(size=len(semnal))) ** 2))
        zgomot = gamma * np.random.normal(size=len(semnal))
        semnal_zgomot = semnal + zgomot

        plt.plot(timp, semnal_zgomot, label=f'Faza: {faza}, SNR: {snr}')

# Setarea detaliilor pentru grafic
plt.title('Semnal sinusoidal cu zgomot la diferite SNR-uri și faze')
plt.xlabel('Timp')
plt.ylabel('Amplitudine')
plt.legend()
plt.show()
```

Exercitiul 3

```python
# Definim parametrii
t = np.linspace(0, 1, 500, endpoint=False)
frecventa1 = 5  # Frecvența primului semnal
frecventa2 = 7  # Frecvența celui de-al doilea semnal

# Generăm cele două semnale
semnal1 = np.sin(2 * np.pi * frecventa1 * t)  # Primul semnal
semnal2 = np.sin(2 * np.pi * frecventa2 * t)  # Al doilea semnal

# Combinăm cele două semnale într-un vector
semnal_combinat = np.concatenate((semnal1, semnal2))

# Afișăm cele două semnale și semnalul combinat
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(t, semnal1, label='Semnal 1 (Freq: 5 Hz)', color='b')
plt.title('Doua semnale cu frecventa diferita')
plt.xlabel('Timp')
plt.ylabel('Amplitudine')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(t, semnal2, label='Semnal2 2 (Freq: 7 Hz)', color='r')
plt.xlabel('Timp')
plt.ylabel('Amplitudine')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(semnal_combinat, label='Semnal Combinat', color='g')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```

Exercitiul 6

```python
# Definim frecvența de eșantionare și timpul
fs = 100  # Frecvența de eșantionare
t = np.linspace(0, 1, fs, endpoint=False)

# Definim frecvențele fundamentale
frecvente = [fs / 2, fs / 4, 0]

# Generăm cele trei semnale sinusoidale
semnale = [np.sin(2 * np.pi * f * t) for f in frecvente]

# Afișăm cele trei semnale
plt.figure(figsize=(10, 6))

for i in range(3):
    plt.subplot(3, 1, i + 1)
    plt.plot(t, semnale[i], label=f'f = {frecvente[i]} Hz')
    plt.title(f'Semnal sinusoidal cu frecvența {frecvente[i]} Hz')
    plt.xlabel('Timp')
    plt.ylabel('Amplitudine')
    plt.legend()

plt.tight_layout()
plt.show()
```

Exercitiul 7

```python
# Definim frecvența de eșantionare și timpul
fs = 1000  # Frecvența de eșantionare inițială
t = np.linspace(0, 1, fs, endpoint=False)

# Generăm semnalul sinusoidal inițial
initial_signal = np.sin(2 * np.pi * 5 * t)  # Presupunem o frecvență de 5 Hz

# Decimăm semnalul
decimated_signal_1 = initial_signal[::4]  # Decimare începând cu primul element
decimated_signal_2 = initial_signal[1::4]  # Decimare începând cu al doilea element

# Afișăm graficele
plt.figure(figsize=(12, 6))

# Afișăm semnalul inițial
plt.subplot(3, 1, 1)
plt.plot(t, initial_signal, label='Initial Signal', color='b')
plt.title('Semnalul inițial și semnalele decimate')
plt.xlabel('Timp')
plt.ylabel('Amplitudine')
plt.legend()

# Afișăm primul semnal decimat
plt.subplot(3, 1, 2)
t_decimated = np.linspace(0, 1, len(decimated_signal_1), endpoint=False)
plt.plot(t_decimated, decimated_signal_1, label='Decimated Signal (Start from 1st element)', color='r')
plt.xlabel('Timp')
plt.ylabel('Amplitudine')
plt.legend()

# Afișăm al doilea semnal decimat
plt.subplot(3, 1, 3)
t_decimated = np.linspace(0, 1, len(decimated_signal_2), endpoint=False)
plt.plot(t_decimated, decimated_signal_2, label='Decimated Signal (Start from 2nd element)', color='g')
plt.xlabel('Timp')
plt.ylabel('Amplitudine')
plt.legend()

plt.tight_layout()
plt.show()
```

(a) În primul grafic, se observă că semnalul inițial are o frecvență de eșantionare de 1000 Hz, în timp ce semnalul decimat are o frecvență de eșantionare de 250 Hz. Astfel, se observă o pierdere semnificativă de informații și o distorsiune a semnalului original.

(b) În al doilea grafic, semnalul decimat începând cu al doilea element arată mai similar cu semnalul inițial decât în cazul decimării începând cu primul element. Acest lucru se datorează faptului că semnalul inițial a fost decimat inițial cu o frecvență de eșantionare de 250 Hz, iar acest proces a eliminat o parte din informațiile semnalului original.
